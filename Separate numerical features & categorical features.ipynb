# Separate numerical features & categorical features

```python
import pandas as pd

df = pd.read_csv('https://www.openml.org/data/get_csv/1595261/adult-census.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
numerical_columns = [
    c for c in df.columns if df[c].dtype.kind in ['i','f']
]
```

```python
data_numeric = df[numerical_columns]
data_numeric.head()
```

```python
categorical_columns = [
    c for c in df.columns if df[c].dtype.kind in ['O']
]
data_categorical = df[categorical_columns]
data_categorical.head()
```
